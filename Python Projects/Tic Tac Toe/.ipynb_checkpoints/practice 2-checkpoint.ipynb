```python
def user_choice():
    choice = 'WRONG'
    acceptable_values=range(0,10)
    
    while choice.isdigit() == False or choice not in acceptable_values:
        
        choice = input("Enter a number (0-10) : ")
        
        if choice.isdigit() == False:
            print("Sorry that's not a digit!!!")
            
        else :
            print("Enter a no. within range : ")
    
    return int(choice)
```
